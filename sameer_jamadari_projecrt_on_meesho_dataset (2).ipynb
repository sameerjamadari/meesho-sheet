```
print("DataFrame Info:")
df.info()

print("\nDataFrame Descriptive Statistics:")
df.describe()

print("\nMissing Values per Column:")
df.isnull().sum()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total sales by product category
category_sales = df.groupby('Product_Category')['Total_Amount'].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis', hue=category_sales.index, legend=False)
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

time_sales = df.groupby('order_date')['Total_Amount'].sum()

plt.figure(figsize=(12, 6))
sns.lineplot(x=time_sales.index, y=time_sales.values)
plt.title('Total Sales Over Time (Daily)')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
import pandas as pd

df = pd.read_excel("/content/Meesho_Orders_Dataset.xlsx")

customer_sales = df.groupby('Customer_ID')['Total_Amount'].sum()
top_10_customers = customer_sales.sort_values(ascending=False).head(10)

print("Top 10 Customers by Sales (using Total_Amount):")
print(top_10_customers)
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_customers.index, y=top_10_customers.values, palette='viridis', hue=top_10_customers.index, legend=False)
plt.title('Top 10 Customers by Total Sales (Total_Amount)')
plt.xlabel('Customer ID')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
average_price_per_category = df.groupby('Product_Category')['Price'].mean()

print("Average Price per Product Category:")
print(average_price_per_category)
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=average_price_per_category.index, y=average_price_per_category.values, palette='viridis', hue=average_price_per_category.index, legend=False)
plt.title('Average Price per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Price (INR)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
